```python
# csv 피일 읽기
df = pd.read_csv("data/cafe.csv")

# 일부만 보기, 위에서/밑에서
# df.head()
# df.tail()

# 통계 요약보기
# df.info()
# df.describe() #숫자만
# df.describe(include="object") #문자열만

# 칼로리 자료형 변환 > 실수
df['칼로리'] = df['칼로리'].astype('float')
print(df['칼로리'].head(2))

# 파생 컬럼 생성
df['할인가'] = (df['가격'] * 0.8).astype('int')
print(df['가격'].head(), df['할인가'].head(), sep='\n')

# df = df.rename(columns={"할인가":"할인가격"})  #변경후 저장 방식
df.rename(columns={"할인가":"할인가격"}, inplace=True) #바로 저장 방식

# print(df.info())

# cafe2.csv 로 저장하기
df.to_csv("data/cafe2.csv", index=False)
```

```python
# 결측치 대체(수치형) - 평균 mean(), 중앙 median()
# 결측치 대체(범주형, 문자형) - 최빈 mode()

import pandas as pd
import numpy  as np

df = pd.DataFrame({
     'A':[1,None,3, 5, 4]
    ,'B':[4,5,None, 7, 8]
    ,'C':[7,None,9, 10, None]
    ,'D':['sky','river','sky', None,'river']
})

# 결측값 대체 - 수치형
# c_val = round(df['C'].mean(), 1)
c_val = df['C'].median()
print('C값:', c_val)
df['C'] = df['C'].fillna(c_val)


# 결측값 대체 - 범주형
d_val = df['D'].mode()
print('D값:\n', d_val, type(d_val))
df['D'] = df['D'].fillna(d_val[0])

df
```
